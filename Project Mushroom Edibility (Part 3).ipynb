## Feature engineering

### One-hot encoding

```python
# Step 1: Import pandas as pd
import pandas as pd
```

```python
# Step 2: Read your file from Part I
df = pd.read_csv('mushroom_data_clean.csv')
df
```

```python
# Step 4: Turn the values in 'class' into binaries
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['class'] = le.fit_transform(df['class'])
df
```

```python
# Step 5: Dummify your DataFrame
df = pd.get_dummies(df, drop_first = True)
df
```

```python
# Step 6: Export DataFrame into a CSV
df.to_csv('mushroom_data_short')
```
